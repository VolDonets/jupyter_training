### Conditional selection (boolean arrays)
We saw conditional selection applied to Series and it'll work in the same way for DataFrames. After all, a DataFrame is a collection of Series.

```python
df
```

```python
df['Population'] > 70
```

```python
df.loc[df['Population'] > 70]
```

```python
df['Population'].mean()
```

```python
df.loc[df['Population'] > 70, 'Population']
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

.

# Modifying DataFrames
It's simple and intuitive, You can add columns, or replace values for columns without issues:

### Adding a new column

```python
langs = pd.Series(['French', 'German', 'Italian'],
                  index=['France', 'Germany', 'Italy'],
                  name='Language'
                 )
langs
```

```python
df['Language'] = langs
```

```python
df
```

### Replacing values per column

```python
df['Language'] = 'English'
```

```python
df
```

### Renaming Columns

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kindom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

### Dropping columns

```python
df.drop(columns='Language', inplace=True)
```

```python
df
```

### Adding values

```python
df.append(pd.Series({
    'Population': 3,
    'GDP': 5
}, name='China'))
```

```python
df
```

```python
df.loc['China']=pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
```

```python
df
```

```python
df.drop('China', inplace=True)
```

```python
df
```

### More radical index changes

```python
df.reset_index()
```

```python
df
```

```python
df.set_index('Population')
```

```python
df
```

## Creating columns from other columns
Altering a DataFrame often involves combining different columns into another. For example, in our Countires analysis, ew could try to calculate the 'GDP per capita', which is just: GDP / Population

```python
df[['Population', 'GDP']]
```

```python
df['GDP'] / df['Population']
```

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```
