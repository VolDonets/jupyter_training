# Reading data from relational databases

```python
import pandas as pd
```

## Read data from SQL database

```python
import sqlite3
```

```python
conn = sqlite3.connect('files/chinook.db')
```

Once we have a 'Connection' object, we can then create a 'Cursor' object. Cursors allow us to execute SQL queries against a databes:

```python
cur = conn.cursor()
```

```python
cur.execute('SELECT * FROM employees LIMIT 5;')
```

```python
result = cur.fetchall()
```

```python
result
```

```python
df = pd.DataFrame(result)
```

```python
df.head()
```

```python
cur.close()
```

```python
conn.close()
```

## USING 'read_sql' METHOD

```python
conn = sqlite3.connect('files/chinook.db')
```

```python
df = pd.read_sql('SELECT * FROM employees;', conn)
```

```python
df.head()
```

```python
df = pd.read_sql('SELECT * FROM employees;', conn,
                index_col='EmployeeId',
                parse_dates=['BirthDate', 'HireDate'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['ReportsTo'].isna().sum()
```

```python
df['ReportsTo'].mean()
```

```python
df['City'] = df['City'].astype('category')

df.info()
```

### Using pandas 'read_sql_quety' method

```python
df = pd.read_sql_query('SELECT * FROM employees LIMIT 5;', conn)
df.head()
```

```python
df = pd.read_sql_query('SELECT * FROM employees LIMIT 5;', conn,
                      index_col='EmployeeId',
                      parse_dates=['BirthDate', 'HireDate'])
df.head()
```

### Using 'read_sql_table' method

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///files/chinook.db')

connection = engine.connect()
```

```python
df = pd.read_sql_table('employees', con=connection)
```

```python
connection.close()
```
